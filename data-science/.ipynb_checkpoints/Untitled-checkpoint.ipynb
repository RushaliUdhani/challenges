```python
import pandas as pd
```

```python
train_data_rms = pd.read_csv("data/train/unit0000_rms.csv")
train_data_alarm= pd.read_csv("data/train/unit0000_alarms.csv", header=None)
```

```python
headers = ['timestamp','message']

train_data_alarm.columns = headers
```

```python
# train_data_rms['Unit'] = '0'
train_data_rms.insert(0, 'Unit', '0')
train_data_alarm.insert(0,'Unit','0')
```

```python
train_data_rms.head()
```

```python
train_data_alarm.head()
```

```python
# format datetime field which comes in as string
train_data_rms['timestamp'] = pd.to_datetime(train_data_rms['timestamp'], format="%Y-%m-%d %H:%M:%S")
train_data_rms.index = pd.to_datetime(train_data_rms.index, unit='s')

print("Total number of rms records: %d" % len(train_data_rms.index))
print(train_data_rms.head())
train_data_rms.describe()
```

```python
# format datetime field which comes in as string
train_data_alarm['timestamp'] = pd.to_datetime(train_data_alarm['timestamp'], format="%Y-%m-%d %H:%M:%S")
train_data_alarm['message'] = train_data_alarm['message'].astype('category')

print("Total number of error records: %d" % len(train_data_alarm.index))
train_data_alarm.head()
```

```python
# Calculate mean values for rms features
temp = []
fields = ['rpm','motor_voltage','motor_current','motor_temp','inlet_temp']
for col in fields:
    temp.append(pd.pivot_table(train_data_rms,
                               index='timestamp',
                               columns='Unit',
                               values=col).resample('3H', closed='left', label='right', how='mean').unstack())
rms_mean_3h = pd.concat(temp, axis=1)
rms_mean_3h.columns = [i + 'mean_3h' for i in fields]
rms_mean_3h.reset_index(inplace=True)

# repeat for standard deviation
temp = []
for col in fields:
    temp.append(pd.pivot_table(train_data_rms,
                               index='timestamp',
                               columns='Unit',
                               values=col).resample('3H', closed='left', label='right', how='std').unstack())
rms_sd_3h = pd.concat(temp, axis=1)
rms_sd_3h.columns = [i + 'sd_3h' for i in fields]
rms_sd_3h.reset_index(inplace=True)

rms_mean_3h.head()
```

```python
# merge columns of feature sets created earlier
rms_feat = pd.concat([rms_mean_3h,
                            rms_sd_3h.ix[:, 2:6],
                          ], axis=1).dropna()
rms_feat.describe()
```

```python
# create a column for each error message type
error_count = pd.get_dummies(train_data_alarm.set_index('timestamp')).reset_index()
error_count.columns = ['timestamp', 'Unit', 'error', 'warning']

# # combine errors for a given machine in a given hour
error_count = error_count.groupby(['Unit', 'timestamp']).sum().reset_index()
error_count.head(13)
```

```python
error_count.dtypes
```

```python
error_count['error']=error_count['error'].astype(int)
error_count['warning']=error_count['warning'].astype(int)
```

```python
error_count.describe()
```

```python
temp = []
fields = ['error','warning']
for col in fields:
    temp.append(error_count.rolling(pd.pivot_table(error_count,
                               index='timestamp',
                               columns='Unit',
                               values=col),window=24).resample( closed='left', label='right', how='first').unstack()).sum()
error_count = pd.concat(temp, axis=1)
error_count.reset_index(inplace=True)
error_count = error_count.dropna()
error_count.describe()
```

```python
train_data_rms['timestamp']=train_data_rms['timestamp'].astype(object)
train_data_alarm['timestamp']=train_data_alarm['timestamp'].astype(object)
error_count['Unit']=error_count['Unit'].astype(object)
error_count['timestamp']=error_count['timestamp'].astype(object)
rms_feat['timestamp']=rms_feat['timestamp'].astype(object)
```

```python
train_data_rms.dtypes
train_data_alarm.dtypes
error_count.dtypes
rms_feat.dtypes
```

```python
error_count.dtypes
```

```python
final_feat = pd.concat([rms_feat, error_count], axis = 1)
```

```python
print(final_feat.head())
final_feat.describe()
```

```python
# Lastly, we merge all the feature data sets we created earlier to get the final feature matrix.

final_feat = rms_feat.merge(error_count, on=['timestamp', 'Unit'], how='left')

print(final_feat.head())
final_feat.describe()
```
